```python

import pandas as pd

# Loading the dataset
data = pd.read_csv(r"C:\Users\ADMIN\Desktop\Kenya_Marriage_Divorce_Dataset (1).csv")

# Displaying summary statistics
summary_stats = data.describe()
print(summary_stats)
```

```python
import matplotlib.pyplot as plt

#  data by year
yearly_data = data.groupby("Year").sum(numeric_only=True)

# Plotting marriages and divorces over time
plt.figure(figsize=(12, 6))
plt.plot(yearly_data.index, yearly_data["Marriages_Registered"], label="Marriages Registered", marker="o")
plt.plot(yearly_data.index, yearly_data["Divorces_Registered"], label="Divorces Registered", marker="o")
plt.title("Trends in Marriages and Divorces in Kenya (2000-2023)")
plt.xlabel("Year")
plt.ylabel("Count")
plt.legend()
plt.grid(True)
plt.show()
v
```

```python
#  data by county
county_data = data.groupby("County").sum(numeric_only=True)

# Sorting counties by the number of marriages
county_data_sorted = county_data.sort_values("Marriages_Registered", ascending=False)

# Plotting bar chart for marriages and divorces by county
county_data_sorted[["Marriages_Registered", "Divorces_Registered"]].plot(kind="bar", figsize=(12, 6), width=0.8)
plt.title("Marriages and Divorces by County")
plt.xlabel("County")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(["Marriages Registered", "Divorces Registered"])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# data by age group
age_group_data = data.groupby("Age_Group").sum(numeric_only=True)

# Plotting bar chart for marriages and divorces by age group
age_group_data[["Marriages_Registered", "Divorces_Registered"]].plot(kind="bar", figsize=(10, 5), color=["skyblue", "salmon"])
plt.title("Marriages and Divorces by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(["Marriages Registered", "Divorces Registered"])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Saving yearly data to a CSV file
yearly_data.to_csv("Yearly_Marriage_Divorce_Trends.csv")
```
